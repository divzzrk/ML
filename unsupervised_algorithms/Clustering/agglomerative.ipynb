```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
x, y = make_blobs()
print('dimensions: ', x.shape)
```

## Performing agglomerative clustering

```python
agg = AgglomerativeClustering(n_clusters=3)
ypred = agg.fit_predict(x)
```

## Creating a dendogram

```python
from scipy.cluster.hierarchy import dendrogram, ward
result = ward(x)
dendrogram(result)
plt.xlabel('Observations')
plt.ylabel('Clusters')
plt.show()
```

```python
import pandas as pd

mydata = {'original': y, 'predicted': ypred}
df = pd.DataFrame(mydata, columns=mydata.keys())
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(df['original'], df['predicted'])
ConfusionMatrixDisplay(confusion_matrix=cm).plot()
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(df['original'], df['predicted'])
print(acc)
```
