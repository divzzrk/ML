# Data Visualisation

## Types of analysis

- 1. Univariate - study of one variable
        Eg: histogram, bar chart, box(1 variables distribution pattern, outliers, can also be multivariate)
- 2. Bivariate - study of two variables
        Eg: line, scatter
- 3. Multivariate - study of more than 2 variables in a single chart
        Eg: piechart, heatmap (to learn the correlation between data, dark(high postive), light(low positive))

# 1.5 IQR Rule (Using Box Plot)
- BOX PLOT is a chart to identify outliers
- Used to identify the outlier
- compute IQR
- whatever is beyond lower level and outer level is outlier
- lower level = Q1 - (1.5 * IQR)
- upper level = Q3 + (1.5 * IQR)


- based on these upper level, lower we can remove outliers

```python
housing_df = pd.read_csv('Datasets/housing_data.csv')
housing_df.head()
```

```python
# how many unique values in sale conditions

housing_df['SaleCondition'].unique()
```
