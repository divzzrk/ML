# Probability Distribution using Scipy

- scipy is used for statistical and probaility related computations
    - package stats has all those computations like bernoulli, poisson, binomial etc

- probability function of discrete variables
    - probability mass function (PMF)
    - bernoulli, poisson, binomial
- probability function of continous variables
    - probability density function (PDF)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

# Bernoulli distribution
if a machine has 70% chance of working
what is the probability of the machine working in a given day?

```python
p = 0.7 #probability the machine will work
x = [0, 1]# choices = either it will work or not work
# since bernoulli (discrete) we need to use PMF

pmf = bernoulli.pmf(x, p)
print(pmf)

#outcome will be two
# not working, working
# probBILITY of getting 0, probability of getting 1
print("Probability of working = ", pmf[1])

plt.bar(x, pmf)
```

# Binomial Distribution

- flip coins 5 times probability of getting 3 heads
- same event happening repeated whats the outcome

```python
from scipy.stats import binom

n = 5 #number_of_trials
p = 0.5 #probability_of_getting_head
k = 3 #trial

pmf = binom.pmf(k, n, p)
print(pmf)
```

# Poisson Distribution
- what is the outcome you will get in a particular time frame
- what is the probability of getting emails in an hour
- imagine u get 3 emails in an hour in average
- what is the probabiity of getting 5 emails in the next one hour

```python
from scipy.stats import poisson

mean_avg = 3
n = 5
pmf = poisson.pmf(n, mean_avg)
print(pmf)
```

# Normal Distribution

- question: probability some one is taller than 180 cm?
- mean = 170
- std_dev = 10 (how far it deviates from center)
- (x - mu)/sigma
- cumulative distribution function (CDF)
- cumulatively taking all values its like he can be 180, 181, 182, till the end of sample (ex if 210 from 181 to 210)

```python
from scipy.stats import norm

mean = 170
std = 10
x = 180

pdf = norm.pdf(x, mean, std)
print(pdf)
probability_above_180 = 1 - norm.cdf(x, mean, std)
print("Probability of people above 180: ", probability_above_180) #15%
```

```python
x = np.linspace(130, 210, 100)
pdf = norm.pdf(x,x.mean(), x.std() )
plt.plot(x, pdf)
```

# Uniform Distribution

```python
from scipy.stats import uniform
a = 10
b = 20
#chance of getting a number less than 15
prob_lt_15 = uniform.cdf(15, loc=a, scale=b-a)
# loc - where it starts from
# scale - range of values
# 15 is the quantile
# here whats prob its 10, 11, 12, 13, 14 and add all and thats the answer
print(prob_lt_15)
```
