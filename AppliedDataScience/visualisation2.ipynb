# binning

## Feature transformation

### trnaforming a numerical to categorical values

```python
housing_df['SalePrice'].describe()
```

```python
price_bins = [0, 45000, 200000, 400000, 755000]
price_labels = ['Class1', 'Class2', 'Class3', 'Class4']

# 0 - 45000 -> class 1
# 45k - 2lakh -> class 2

housing_df['Price Range'] = pd.cut(housing_df["SalePrice"], bins=price_bins, labels=price_labels)
```

```python
housing_df["Price Range"].unique()
```

```python
housing_df["Price Range"].value_counts()
```

```python
price_counts = housing_df["Price Range"].value_counts()
plt.pie(price_counts, labels=price_counts.index)
```

# Step 2: handling missing data

```python
movie_df.isnull().sum()
```

```python
# lets take gross as an example as its the highest null

movie_df["gross"].fillna(movie_df["gross"].mean, inplace=True)
```

```python
movie_df["gross"].isnull().sum()
```

```python
# now take a categorical variable and check
movie_df['color'].unique()
```

```python
movie_df["color"].value_counts()
```

```python
movie_df['color'].mode() # so if there are 2 values mod returns 2 so take the first
```

```python
movie_df["color"].fillna(movie_df["color"].mode()[0], inplace=True)
```

```python
movie_df['color'].unique()
```
